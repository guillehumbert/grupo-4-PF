```python
# >>> VISUALIZACION <<<
#Importa las librerias matplotlib y seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# visualizando la data en gráfico de barra
male_female = hap['Sex'].value_counts() #cuantas veces aparece cada categoria
plt.bar(male_female.index, male_female.values) # labels y values?
plt.xlabel('Category') 
plt.ylabel('Count')
plt.title('Male vs Female')
plt.show() #muestra el grafico
```

```python
# paises con less heart attacks 
# 1. contar la cantidad de paises totales para mostrar en y
# 2. sumar la cantidad de heart attack risk por country 
# 3. mostrar los paises que tienen menores valores en un top 5

count_country = hap["Country"].value_counts()
sum_har = hap.groupby("Country")["Heart Attack Risk"].sum()
top_countries = sum_har.nsmallest(5) # muestra el top 5 valores mín

plt.bar(top_countries.index,top_countries.values)
plt.xlabel("Countries")
plt.ylabel("Count")
plt.title("Least Heart Attack Risk")
plt.show()
```
